```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
import numpy as np

# Add new target for multi-class: 2=Win, 1=Draw, 0=Loss
def get_result_code(result):
    if result == "W":
        return 2
    elif result == "D":
        return 1
    else:
        return 0

matches["target_multi"] = matches["result"].apply(get_result_code)
```

```python
# Recent form: points in last 5 matches
def recent_form(group):
    group = group.sort_values("date")
    points = group["result"].map({"W": 3, "D": 1, "L": 0})
    group["recent_points"] = points.rolling(5, closed='left').sum()
    return group

matches = matches.groupby("team").apply(recent_form).reset_index(drop=True)
```

```python
# Home/Away win rate (last 10 home/away games)
def home_away_win_rate(group):
    group = group.sort_values("date")
    group["home_win_rate"] = (
        (group["venue"] == "Home") & (group["result"] == "W")
    ).rolling(10, closed='left').mean()
    group["away_win_rate"] = (
        (group["venue"] == "Away") & (group["result"] == "W")
    ).rolling(10, closed='left').mean()
    return group

matches = matches.groupby("team").apply(home_away_win_rate).reset_index(drop=True)
```

```python
# Head-to-head win rate (last 5 matches vs opponent)
def h2h_win_rate(row):
    team = row["team"]
    opp = row["opponent"]
    date = row["date"]
    prev_matches = matches[
        ((matches["team"] == team) & (matches["opponent"] == opp)) &
        (matches["date"] < date)
    ].sort_values("date").tail(5)
    if prev_matches.empty:
        return np.nan
    return (prev_matches["result"] == "W").mean()

matches["h2h_win_rate"] = matches.apply(h2h_win_rate, axis=1)
```

```python
# Update rolling averages as before
matches["date"] = pd.to_datetime(matches["date"])
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
matches_rolling = matches_rolling.droplevel('team')
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
# Merge new features into matches_rolling
for feat in ["recent_points", "home_win_rate", "away_win_rate", "h2h_win_rate"]:
    matches_rolling[feat] = matches.loc[matches_rolling.index, feat].values
matches_rolling["target_multi"] = matches.loc[matches_rolling.index, "target_multi"].values
```

```python
# Define predictors for multi-class
predictors_multi = predictors + new_cols + ["recent_points", "home_win_rate", "away_win_rate", "h2h_win_rate"]
```

```python
# Drop rows with missing values in new features
matches_rolling_multi = matches_rolling.dropna(subset=predictors_multi + ["target_multi"])

# Train/test split
train = matches_rolling_multi[matches_rolling_multi["date"] < "2022-01-01"]
test = matches_rolling_multi[matches_rolling_multi["date"] >= "2022-01-01"]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

rf_multi = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
rf_multi.fit(train[predictors_multi], train["target_multi"])
preds_multi = rf_multi.predict(test[predictors_multi])
```

```python
# Evaluation
print("Classification report (Win=2, Draw=1, Loss=0):")
print(classification_report(test["target_multi"], preds_multi, target_names=["Loss", "Draw", "Win"]))

print("Confusion matrix:")
print(confusion_matrix(test["target_multi"], preds_multi))
```

```python
def predict_specific_match_multi(match_index):
    row = matches_rolling_multi.iloc[match_index]
    X = pd.DataFrame([row[predictors_multi].values], columns=predictors_multi)
    predicted = rf_multi.predict(X)[0]
    actual = row["target_multi"]
    result_map = {2: "Win", 1: "Draw", 0: "Loss"}
    print(f"Match info: {row['team']} vs {row['opponent']} on {row['date'].date()}")
    print(f"Predicted outcome: {result_map[predicted]}")
    print(f"Actual outcome: {result_map[actual]}")
    print(f"Result column: {row['result']}")
```
